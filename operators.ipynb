```python
## Arthmetic Operators
print(5+6)
print(5-6)
print(5*6)
print(5/6)
print(6//5) ## // is a integer division convert decimal into integer
print(5%2) ## % modulus operator give reminder
print(5**2) ## ** means power
```

```python
## Relational Operators
print(4>5)
print(4<5)
print(4>=5)
print(4<=5)
print(4==4)
print(4!=4)
```

```python
## Logical Operators (AND,OR,NOT)
print( 1 and 0)
print( 1 or 0)
print(not 1)
```

```python
## Bitwise Operators

## bitwise And
print(2 & 3) ## & for And

## Bitwise Or
print(2 | 3) ## | for or

## Bitwise XOR
print(2 ^ 3) ## ^ For XOR

print(~3) ## bitwise not

print(4 >> 2) ###  bitwise left

print(4 << 2) ###  bitwise Right

```

```python
## Assignment Operators(Assign a value to variable)

a = 2

a += 2 ## (a = a +2 )

print(a)
```

```python
## Membersship Operator
 ## in/not in
print('D'  in 'Delhi')

print(1 in [3,4,5,6,7,8])
```
